```python
# Create Outcome
data.loc[data.y == "yes", "outcome"] = 1
data.loc[data.y == "no", "outcome"] = 0

x = pd.get_dummies(data[["loan","contact","day_of_week","poutcome","marital","education","default","month"]])
x["duration"] = data["duration"]
x["campaign"] = data["campaign"]
x["pdays"] = data["pdays"]
x["previous"] = data["previous"]
y = data["outcome"]

x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=22)
```

```python
def print_model_output(input_model):
    model=input_model
    model.fit(x_train,y_train)
    train_score=model.score(x_train,y_train)
    test_score=model.score(x_test,y_test)
    predict_test=model.predict(x_test)
    table_test=pd.crosstab(y_test, predict_test,margins=True)
    print("Train Score %0.6f" % (train_score))
    print("Test Score %0.6f" % (test_score)) 
    print("Cross-Val-Score:")
    print(cross_val_score(model,x_train,y_train))
```

```python
#Random Forest
estimators= (10,100,500,1000)
for estimator in estimators:
    print("Estimator: %0.f" %(estimator))
    print_model_output(ensemble.RandomForestClassifier(n_estimators=1,max_features="auto"))
```

```python
print_model_output(KNeighborsClassifier())
```

```python
def boosted_model(loss,estimators):
    model = ensemble.GradientBoostingClassifier(loss=loss,n_estimators=estimators)
    return model
```

```python
#Boosted Model
loss_types=("exponential","deviance")
estimators_list=(100,250,500)
for (loss,estimators) in itertools.product(loss_types,estimators_list):
    print(loss)
    print("Estimators: %0.f" % (estimators))
    print_model_output(boosted_model(loss,estimators))
```

```python
#Logistic Regression
print_model_output(LogisticRegression(C=1e09,penalty="l2"))
```

```python
#Ridge Regression
print_model_output(LogisticRegression(C=1,penalty="l2"))
```

```python
#Lasso Regression
print_model_output(LogisticRegression(C=1,penalty="l1"))
```
